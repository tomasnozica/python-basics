## Ciclos

Los ciclos o iteraciones son maneras de realizar un bloque de codigo n veces. Esas n veces puede ser recorriendo alguna estructura o mientras se cumpla una condicion.

### While

La sintaxis es la siguiente:

While *(condicion)*:
    
    *hacer algo*

```python
vuelta = 0
total = 5
while vuelta < total:
    print("Voy por la vuelta: ",vuelta)
    vuelta+=1
```

Como es el funcionamiento paso a paso del while:

1. Evalua la condicion, en nuestro caso se pregunta ¿vuelta < total?
2. Como vuelta inicialmente es igual a 0, total igual a 5 y 0 < 5: devuelve True
3. Como dio True la condicion, ejecuta el codigo debajo del while
    
    a. En un primer momento imprime en que vuelta se encuentra
    
    b. vuelta+=1 lo que hace es sumar 1 al valor vuelta, por lo cual en la primer vuelta pasa de 0 a 1

4. Vuelve a evaluar la condicion, como vuelta ahora es 1, total sigue igual a 5 y 1 < 5: devuelve True.

El bloque se sigue ejecutando. Llega un punto en que vuelta = 4. Veamos el paso a paso.

1. Evalua la condicion, en nuestro caso se pregunta ¿vuelta < total?
2. Como vuelta ahora es igual a 4, total igual a 5 y 4 < 5: devuelve True
3. Como dio True la condicion, ejecuta el codigo debajo del while
    
    a. En un primer momento imprime en que vuelta se encuentra
    
    b. vuelta+=1 lo que hace es sumar 1 al valor vuelta, por lo cual en la primer vuelta pasa de 4 a 5
    
4. Vuelve a evaluar la condicion, como vuelta ahora es 5, total sigue igual a 5 y 5 < 5: devuelve False.

Como dio False, no ejecuta el codigo y sigue la ejecucion del programa o termina, segun corresponda

```python
vuelta = 0
total = 5
while vuelta < total:
    print("Voy por la vuelta: ",vuelta)
    vuelta+=1
    
print("Ya estoy fuera del ciclo")
```

### For

El ciclo for tiene una sintaxis distinta pero una logica muy similar

for i in lista:
    
    *hace algo*
    
La diferencia esta en que ese i puede ser usado en el bloque de codigo y su valor cambia con cada iteracion.

```python
for i in "fernando":
    print(i)
```

```python
palabra=''
for letra in "fernando":
    mayus = ord(letra)-32
    palabra+=chr(mayus)
print(palabra)
```

Veamos como opera

En la primer vuelta, i toma el valor 'f'

Con el valor de i en char, lo transformamos a int segun ASCII y lo convertimos en la representacion en mayuscula
Luego a la variable palabra le almacenamos el valor en char.

'palabra' al finalizar la primer vuelta es F

Y asi, sucesivamente llegamos hasta finalizar el recorrido de la palabra.

La encriptacion de mensajes es algo tan viejo como el hombre mismo, hay registros historicos que comentan que el jefe del Estado Romano, Julio Cesar (100 a.C. - 44 a.C) para comunicarse con su ejercito empleaba una tecnica de desplazamientos donde se lograba ocultar el contenido original del mensaje.

Esta técnica es conocida como el "Cifrado Cesar", muy creativo el nombre... Se sabe que consistía en 3 desplazamientos. Por Ejemplo.

    Dado nuestro abecedario: A,B,C,..,Z
    Se remplaza cada caracter por el nuevo abecedario: D,E,F,...,Z,A,B,C.

    El mensaje 'Hola' seria remplazado por

    H O L A
    K R O D

    El mensaje 'Zorzal' seria remplazado por

    Z O R Z A L
    C R U C D O

Nuestro objetivo va a ser crear la funcion **encriptar(frase)** que realice dichos desplazamientos y nos devuelva un mensaje encriptado. Para ello tener en cuenta los mecanismos vistos con la tabla ASCII. *Tener en cuenta la transformacion de una palabra que contenga X, Y o Z. Cuando se lea un espacio, este no debe ser desplazado.*

Ahora que tenemos nuestra funcion que encripta mensajes debemos chequear que la funcion haga lo que le pedimos correctamente. Crear entonces la funcion, **desencriptar(msj_encriptado)**.



```python
def encriptar(frase):
    msj_encriptado = ''
    for letra in frase:
        if letra == 'x':
            msj_encriptado+='a'
        elif letra == 'y':
            msj_encriptado+='b'
        elif letra == 'z':
            msj_encriptado+='c'
        elif letra == ' ':
            msj_encriptado+=' '
        else:
            despl = ord(letra)+3
            msj_encriptado+=chr(despl)
    return msj_encriptado
```

```python
encriptar("hola zorzal")
```

```python
def desencriptar(msj_encriptado):
    frase = ''
    for letra in msj_encriptado:
        if letra == 'a':
            frase+='x'
        elif letra == 'b':
            frase+='y'
        elif letra == 'c':
            frase+='z'
        elif letra == ' ':
            frase+=' '
        else:
            despl = ord(letra)-3
            frase+=chr(despl)
    return frase
```

Cuenta la leyenda que en 1786, en una clase de Aritmética de tercero de primaria, un maestro rural llamado Büttner pidió a sus alumnos que hallaran la suma de los 100 primeros números. Un alumno de esa clase llamado Carl Friedrich Gauss, que entonces tenía 9 años, halló la respuesta correcta en muy poco tiempo, diciendo «Ligget se’» («ya está»). 

Al acabar la hora se comprobaron las soluciones y se vio que la solución de Gauss era correcta, mientras que no lo eran muchas de las de sus compañeros.

### ¿Como hizo Gauss?

Pensó: tengo que sumar los numeros del 1 al 100

S1 = 1 + 2 + 3 + ... + 98 + 99 + 100

Gauss dijo, bueno la suma es conmutativa entonces:

S2 = 100 + 99 + 98 + ... + 3 + 2 + 1

Y si sumo S1 + S2 se da la coincidencia que:

S1 + S2 = (1+100) + (2+99) + (3+98) + ... + (98+3) + (99+2) + (100+1)

es decir,

S1 + S2 = 101 + 101 + 101 + ... + 101 + 101 + 101 (100 veces sumado 101)

ENTONCES: Como la suma de 2 veces la secuencia de numeros del 1 al 100 es 100 veces sumar 101. Escrito lindo:
                   
                                       2 * SUMA(1,...,100) = 100 * 101
                                       
Concluyo que:

                                        SUMA(1,...,100) = (100*101)/2

Como no somos Gauss, pero si sabemos programar, calculemos la "Suma de Gauss" para un arreglo de 100 numeros.

```python
def crear_lista_numeros(inicio, fin):
    lista = []
    while inicio<=fin:
        lista.append(inicio)
        inicio+=1
    return lista
```

```python
lista = crear_lista_numeros(1,100)
```

```python
def suma_de_gauss(lista):
    resultado = 0
    for numero in lista:
        resultado+=numero
    return resultado
```

```python
suma_de_gauss(lista)
```

Chequeemos si Gauss tenia razon:

```python
suma_de_gauss(lista) == (100*101)/2
```
